```python
import csv
import math
import numpy as np
import random
import pylab
import matplotlib.pyplot as plt
with open('Datatest-100000.csv', newline='') as f:
    data = list(csv.reader(f))
```

```python
print(data)
```

```python
time = np.genfromtxt('Datatest-100000.csv', delimiter=',',names=True)
x = time['N']
y = time['Milliseconds']
```

```python
plt.scatter(x,y)
plt.xlabel('N')
plt.ylabel('Milliseconds')
plt.title('Total Running Time')
plt.show()
```

```python
# data[0] = ['0', '150']
data = data[1:]
```

```python
data = np.array(data,dtype=float)
```

```python
x = data[:,0]
y = data[:,1]
plt.plot(x, np.log(y))
plt.show()
```

```python
p = np.poly1d(np.polyfit(x, np.log(y), 1))
t = np.linspace(0, 100000, 1000)
plt.plot(x, y, 'o', t, np.exp(p(t)), '-')
plt.title("exponential function")
plt.show()
```

```python
p = np.poly1d(np.polyfit(x, y, 3))
t = np.linspace(0, 100000, 1000)
plt.plot(x, y, 'o', t, p(t), '-')
plt.title("quadratic function")
plt.show()
```

```python
p = np.poly1d(np.polyfit(x, y, 2))
t = np.linspace(0, 100000, 1000)
plt.plot(x, y, 'o', t, p(t), '-')
plt.title("cubic function")
plt.show()
```
